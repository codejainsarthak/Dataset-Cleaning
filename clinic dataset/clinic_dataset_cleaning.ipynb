```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\PYTHON\Graphs_matplotlib\seaborn\dataset cleaning\clinic dataset\messy_medical_clinic_dataset.csv")

df = pd.DataFrame(data)

print(df.head(20))
```

```python
print(df.duplicated().sum())
df = df.drop_duplicates()
```

```python
print(df["patient_id"].unique())

df["patient_id"] = df["patient_id"].replace(r"[^0-9]" , "", regex= True)

print(df["patient_id"].head(20))

df["patient_id"] = pd.to_numeric(df["patient_id"])
```

```python
print(df["patient_name"].unique())
df["patient_name"] = df["patient_name"].str.lower()

df["patient_name"] = df["patient_name"].replace("???" , np.nan)

df["patient_name"] = df["patient_name"].str.capitalize()

print(df["patient_name"].head(20))
```

```python
# df["age"].unique()
age_map = {"unknown" :np.nan,
           "thirty" : 30}

df["age"] = df["age"].replace(age_map)
df["age"] = pd.to_numeric(df["age"])
```

```python
df["gender"].unique

df["gender"] = df["gender"].str.lower()

gender_mapp = {"f" : "female" , 
               "m" : "male"}

df["gender"] = df["gender"].replace(gender_mapp)

df["gender"] = df["gender"].str.upper()

# print(df["gender"].head(20))

df["gender"] = df["gender"].str.strip()

df["gender"] = df["gender"].replace("unknown" , np.nan)
print(df["gender"].head(20))
```

```python
df.columns

df["blood_group"].unique()

df["blood_group"] = df["blood_group"].replace("unknown" , np.nan)
df["blood_group"] = df["blood_group"].str.strip()
df["blood_group"] = df["blood_group"].str.upper()
print(df["blood_group"].head(20))
```

```python
df.columns
df["weight"].unique()
df["weight"] = df["weight"].replace("unknown" , np.nan)
df["weight"] = df["weight"].replace(r"[^0-9.]" , "", regex = True)
df["weight"] = df["weight"].str.strip()
df["weight"] = pd.to_numeric(df["weight"])

print(df["weight"].head(20))
```

```python
df["height"].unique()
df["height"] = df["height"].replace("error" , np.nan)
df["height"] = df["height"].replace(r"[^0-9.]" , "", regex = True)
df["height"] = df["height"].str.strip()
df["height"] = pd.to_numeric(df["height"])
```

```python

df["diagnosis"].unique()
df["diagnosis"] = df["diagnosis"].replace("???" , np.nan)
df["diagnosis"] = df["diagnosis"].str.strip()
df["diagnosis"] = df["diagnosis"].str.upper()
```

```python
# df["consultation_fee"].unique()

df["consultation_fee"] = df["consultation_fee"].replace("missing" , np.nan)
df["consultation_fee"] = df["consultation_fee"].replace(r"[^0-9.]" , "" , regex = True)
df["consultation_fee"] = pd.to_numeric(df["consultation_fee"])
```

```python
df["visit_date"].unique()
df["visit_date"] = df["visit_date"].replace("not available" , np.nan)
df["visit_date"] = pd.to_datetime(df["visit_date"] , errors = "coerce")
```

```python
print(df["payment_method"].unique())
df["payment_method"] = df["payment_method"].replace("unknown" , np.nan)
df["payment_method"] = df["payment_method"].str.upper()
df["payment_method"] = df["payment_method"].str.strip()
```

```python
# df["phone"].unique()
df["phone"] = df["phone"].replace(r"[^0-9]" , "" , regex = True)
phone_map = {"invalid" : np.nan,
             "unknown" : np.nan}

df["phone"] = df["phone"].replace(phone_map)

df.loc[df["phone"].str.len() > 10 , "phone"] = np.nan
df["phone"] = pd.to_numeric(df["phone"])
```

```python
df["age"] = df["age"].fillna(df["age"].median())
df["weight"] = df["weight"].fillna(df["weight"].median())
df["height"] = df["height"].fillna(df["height"].median())
df["consultation_fee"] = df["consultation_fee"].fillna(df["consultation_fee"].median())

df["blood_group"] = df["blood_group"].fillna(df["blood_group"].mode()[0])
df["diagnosis"] = df["diagnosis"].fillna(df["diagnosis"].mode()[0])
df["gender"] = df["gender"].fillna(df["gender"].mode()[0])

df["payment_method"] = df["payment_method"].fillna(df["payment_method"].mode()[0])
```

```python
df.to_csv("Clinic_Dataset_cleaned.csv" , index = False)
```
