```python
import pandas as pd
import numpy as np

data = pd.read_csv(r"C:\Users\hp\OneDrive\Desktop\PYTHON\Graphs_matplotlib\seaborn\dataset cleaning\Urban Dataset\messy_urban_transport_dataset.csv")

df = pd.DataFrame(data)

print(df.head(20))
```

```python
df.duplicated().sum()
df= df.drop_duplicates()
```

```python
print(df["passenger_name"].unique())
df["passenger_name"] = df["passenger_name"].str.strip()
df["passenger_name"] = df["passenger_name"].str.upper()

print(df["passenger_name"].head(20))
```

```python
df.columns
```

```python
print(df["age"].unique())
df["age"] = df["age"].replace("twenty five" , np.nan)
df["age"] = pd.to_numeric(df["age"])
```

```python
print(df["city"].unique())
df["city"] = df["city"].str.strip()
df["city"] = df["city"].str.upper()
```

```python
df["transport_mode"].unique()
df["transport_mode"] = df["transport_mode"].str.strip()
df["transport_mode"] = df["transport_mode"].str.upper()
```

```python
df["distance_km"].unique()

df["distance_km"] = df["distance_km"].replace("unknown" , np.nan)
df["distance_km"] = pd.to_numeric(df["distance_km"])
```

```python
df["fare"].unique()
df["fare"] = df["fare"].replace("error" , np.nan)
df["fare"] = df["fare"].replace(r"[^0-9.]" , "" ,regex = True)
df["fare"] = pd.to_numeric(df["fare"])
```

```python
df["trip_start"].unique()
df["trip_start"] = pd.to_datetime(df["trip_start"] , errors = "coerce",dayfirst= True)
print(df["trip_start"].head(20))
```

```python
df["trip_end"].unique()
df["trip_end"] = pd.to_datetime(df["trip_end"] , errors = "coerce")
```

```python
df["rating"].unique()
mapp = {"five" :5,
        "-1" : 0,
        }

df["rating"] = df["rating"].replace(mapp)
df["rating"] = pd.to_numeric(df["rating"])
df["rating"].head(20)
```

```python
df["weather_condition"].unique()
df["weather_condition"] = df["weather_condition"].str.strip()
df["weather_condition"] = df["weather_condition"].str.upper()
```

```python
df["payment_method"].unique()
df["payment_method"] = df["payment_method"].str.strip()
df["payment_method"] = df["payment_method"].str.upper()
```

```python
# filling the value of nan

df["age"] = df["age"].fillna(df["age"].median())
df["city"] = df["city"].fillna(df["city"].mode()[0])
df["transport_mode"] = df["transport_mode"].fillna(df["transport_mode"].mode()[0])
df["distance_km"] = df["distance_km"].fillna(df["distance_km"].median())
df["fare"] = df["fare"].fillna(df["fare"].median())
df["rating"] = df["rating"].fillna(df["rating"].median())
df["weather_condition"] = df["weather_condition"].fillna(df["weather_condition"].mode()[0])
df["payment_method"] = df["payment_method"].fillna(df["payment_method"].mode()[0])

```
