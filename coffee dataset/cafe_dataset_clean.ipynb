```python
import pandas as pd
import numpy as np
data = pd.read_csv(r"c:\Users\hp\Downloads\archive\dirty_cafe_sales.csv")
df = pd.DataFrame(data)

print(df.info())
print(df["Location"].head(10))
print(df.shape)
```

```python
# cleaning item columns

print(df["Item"].unique())

mapp = {
    "UNKNOWN" : np.nan,
    "ERROR" : np.nan,

}

df["Item"] = df["Item"].replace(mapp)
df["Item"] = df["Item"].str.upper()

print(df["Item"].head(20))
```

```python
print(df["Quantity"].unique())


df["Quantity"] = df["Quantity"].replace(mapp)
df["Quantity"] = pd.to_numeric(df["Quantity"])
```

```python
print(df["Price Per Unit"].unique())

df["Price Per Unit"] = df["Price Per Unit"].replace(mapp)
df["Price Per Unit"] = pd.to_numeric(df["Price Per Unit"])
```

```python
print(df["Total Spent"].unique())
df["Total Spent"] = df["Total Spent"].replace(mapp)
df["Total Spent"] = pd.to_numeric(df["Total Spent"])
```

```python
print(df["Payment Method"].unique())
df["Payment Method"] = df["Payment Method"].replace(mapp)

```

```python
df["Location"] = df["Location"].replace(mapp)
```

```python
print(df["Price Per Unit"].unique())

df["Price Per Unit"] = df["Price Per Unit"].fillna(df["Price Per Unit"].median())
```

```python
df["Total Spent"] = df["Quantity"] * df["Price Per Unit"]
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"] , errors = "coerce")
```

```python
# df = df.drop(columns = ["Qantity"])
```

```python
## filling the data

print(df["Item"].isna().sum())

df["Item"] = df["Item"].fillna(df["Item"].mode()[0])
df["Quantity"] = df["Quantity"].fillna(df["Quantity"].median())
# 
df["Payment Method"] = df["Payment Method"].fillna(df["Payment Method"].mode()[0])
df["Location"] = df["Location"].fillna(df["Location"].mode()[0])
```

```python
print(df.isna().sum())
```

```python
df.to_csv("Cafe_Sales_Cleaned.csv" , index = False)
```
